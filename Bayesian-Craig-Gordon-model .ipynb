# Bayesian-Craig-Gordon Model for Estimating E/I Ratio in Aquatic Systems


## Mass Balance of a Waterbody

For a waterbody at a hydrologic steady state (kept at a constant volume while undergoing evaporation with liquid outflow),

\[
\frac{dV}{dt} = I - Q - E = 0
\]

Where:
- \( V \) = Volume of the water body
- \( I \) = Rate of inflow from various sources
- \( E \) = Rate of evaporation
- \( Q \) = Rate of outflow


## Isotope Mass Balance 

The isotope mass balance equation using stable water isotope tracer (\( \delta^{18}O \) or \( \delta^2H \)) is given by:

\[
\frac{dV\delta}{dt} = I\delta_I - Q\delta_Q - E\delta_E = 0
\]

Where:
- \( \delta_I \) = Isotopic composition of inflow
- \( \delta_Q \) = Isotopic composition of outflow
- \( \delta_E \) = Isotopic composition of evaporation flux
- \( \delta \) = Isotopic composition of water undergoing evaporation (in ‰)


## Evaporation to Inflow Ratio (E/I)

Using the equations from the mass balance and isotope mass balance, the Evaporation to Inflow ratio (E/I) is defined as:

\[
\frac{E}{I} = \frac{\delta_I - \delta_Q}{\delta_E - \delta_Q}
\]


## Isotopic Composition of Evaporation Flux (\( \delta_E \))

The isotopic composition of the evaporation flux, \( \delta_E \), can be expressed in terms of various parameters related to isotopic composition, humidity, and fractionation factors:

\[
\delta_E = \frac{1}{{1-h + 10^{-3} \cdot \epsilon_k}} \left( \frac{\delta_W - \epsilon^+}{\alpha^+} - h\delta_A - \epsilon_k \right)
\]

And alternatively,

\[
\delta_E = \frac{1}{{1-h + 10^{-3} \cdot \epsilon_k}} \left( \frac{\delta_W}{\alpha^+} - h\delta_A - \epsilon \right)
\]

Where:
- \( \delta_W \) = Measured isotopic composition of the surface water
- \( h \) = Measured (or modeled/assumed) relative humidity normalized to the surface temperature
- \( \epsilon \) = Total fractionation factor, \( \epsilon = \epsilon_k + \frac{\epsilon^+}{\alpha^+} \)
- \( \epsilon_k \) = Kinetic isotope enrichment factor, given by \( \epsilon_k = (1-h) \times C_k \).
\( C_k \) is the kinetic fractionation constant determined experimentally (14.2‰ for \( \delta^{18}O \) and 12.5‰ for \( \delta^2H \)).
- \( \epsilon^+ \) = Equilibrium isotopic separation, \( \epsilon^+ = (\alpha^+ - 1) \times 1000 \)
- \( \alpha^+ \) = Equilibrium isotope fractionation factor, which varies with temperature

## Estimation of Evaporation to Inflow ratio (E/I)

The Evaporation to Inflow ratio, \( \frac{E}{I} \), can be obtained using the isotopic compositions:

\[
\frac{E}{I} = \frac{\delta_I - \delta_Q}{(\delta_Q - \delta^*)m} \
\]

Where:
- \( \delta_I \) = Isotopic composition of inflow
- \( \delta_Q \) = Isotopic composition of outflow
- \( \delta^* \) = Limiting isotope enrichment factor
- \( m \) = Enrichment slope or calculation factor


Here, \( \delta^* \) and \( m \)

The enrichment slope or calculation factor, \( m \), is given by:

\[
m = \frac{h - \frac{\varepsilon}{1000}}{1 - h + \frac{\varepsilon_k}{1000}}
\]

Where:
- \( h \) = Measured (or modeled) relative humidity normalized to the surface temperature.
- \( \varepsilon \) = Total fractionation factor.
- \( \varepsilon_k \) = Kinetic isotope enrichment factor.


The limiting isotope enrichment factor, \( \delta^* \), is given by:

\[
\delta^* = \frac{h \cdot \delta_A + \varepsilon}{h - \frac{\varepsilon}{1000}}
\]

Where:
- \( h \) = Measured (or modeled) relative humidity normalized to the surface temperature.
- \( \delta_A \) = Isotopic composition of ambient atmospheric moisture.
- \( \varepsilon \) = Total fractionation factor.

\[

\varepsilon = \varepsilon_k + \frac{\varepsilon^+}{\alpha^+}

Where:

Here Temp is temperature in celsius and H is Humidity

```python
import numpy as np

# Define the equation to compute alpha+ for d18O
def compute_alpha_plus_d18O(temp_celsius):
    T = temp_celsius + 273.15  # Convert temperature from Celsius to Kelvin
    ln_alpha_plus_d18O = -7.685 + 6.7123 * (10**3) / T - 1.6664 * (10**6) / T**2 + 0.35041 * (10**9) / T**3
    return np.exp(ln_alpha_plus_d18O / 1000)

# Define the equation to compute alpha+ for d2H
def compute_alpha_plus_d2H(temp_celsius):
    T = temp_celsius + 273.15
    ln_alpha_plus_d2H = (1158.8 * (T**3) / (10**9)) - (1620.1 * (T**2) / (10**6)) + (794.84 * T / (10**3)) - 161.04 + (2.9992 * (10**9) / T**3)
    return np.exp(ln_alpha_plus_d2H / 1000)

# Define the equation to compute epsilon+ for d18O and d2H
def compute_epsilon_plus_d18O(temp_celsius):
    return (compute_alpha_plus_d18O(temp_celsius) - 1) * 1000

def compute_epsilon_plus_d2H(temp_celsius):
    return (compute_alpha_plus_d2H(temp_celsius) - 1) * 1000

# Constants for ε_k calculation
C_k_d18O = 14.2  # for δ18O
C_k_d2H = 12.5  # for δ2H

def compute_epsilon_k_d18O(humidity):
    return (1 - humidity) * C_k_d18O

def compute_epsilon_k_d2H(humidity):
    return (1 - humidity) * C_k_d2H

# Compute all values
data['alpha_plus_d18O'] = data['Temp'].apply(compute_alpha_plus_d18O)
data['alpha_plus_d2H'] = data['Temp'].apply(compute_alpha_plus_d2H)
data['epsilon_plus_d18O'] = data['Temp'].apply(compute_epsilon_plus_d18O)
data['epsilon_plus_d2H'] = data['Temp'].apply(compute_epsilon_plus_d2H)
data['epsilon_k_d18O'] = data['H'].apply(compute_epsilon_k_d18O)
data['epsilon_k_d2H'] = data['H'].apply(compute_epsilon_k_d2H)
data['epsilon_d18O'] = data['epsilon_k_d18O'] + data['epsilon_plus_d18O'] / data['alpha_plus_d18O']
data['epsilon_d2H'] = data['epsilon_k_d2H'] + data['epsilon_plus_d2H'] / data['alpha_plus_d2H']

data[['Temp', 'H', 'alpha_plus_d18O',  'alpha_plus_d2H','epsilon_plus_d18O', 'epsilon_plus_d2H', 'epsilon_k_d18O', 'epsilon_k_d2H', 'epsilon_d18O', 'epsilon_d2H']]
```

```python
# Compute δ_A for δ18O and δ2H
data['delta_A_d18O'] = (data['d18O_Rain'] - data['epsilon_plus_d18O']) / data['alpha_plus_d18O']
data['delta_A_d2H'] = (data['d2H_Rain'] - data['epsilon_plus_d2H']) / data['alpha_plus_d2H']

data[['d18O_Rain', 'delta_A_d18O', 'd2H_Rain', 'delta_A_d2H']]
```

```python
# Compute δ* for δ18O and δ2H
data['delta_star_d18O'] = (data['H'] * data['delta_A_d18O'] + data['epsilon_d18O']) / (data['H'] - data['epsilon_d18O'] / 1000)
data['delta_star_d2H'] = (data['H'] * data['delta_A_d2H'] + data['epsilon_d2H']) / (data['H'] - data['epsilon_d2H'] / 1000)

data[['H', 'delta_A_d18O', 'epsilon_d18O', 'delta_star_d18O', 'delta_A_d2H', 'epsilon_d2H', 'delta_star_d2H']]
```

```python
# Compute m for δ18O and δ2H
data['m_d18O'] = (data['H'] - data['epsilon_d18O'] / 1000) / (1 - data['H'] + data['epsilon_k_d18O'] / 1000)
data['m_d2H'] = (data['H'] - data['epsilon_d2H'] / 1000) / (1 - data['H'] + data['epsilon_k_d2H'] / 1000)

data[['H', 'epsilon_d18O', 'epsilon_k_d18O', 'm_d18O', 'epsilon_d2H', 'epsilon_k_d2H', 'm_d2H']]
```

```python
# Assuming δ_Q is approximately equal to the isotopic composition of the water body (δ_W) as given in the context
# Using the columns 'd18O' and 'd2H' from the data as δ_W for δ18O and δ2H respectively.

# Compute E/I for δ18O
data['E_I_d18O'] = (data['d18O_I'] - data['d18O_Q']) / ((data['d18O_Q'] - data['delta_star_d18O']) * data['m_d18O'])

# Compute E/I for δ2H
data['E_I_d2H'] = (data['d2H_I'] - data['d2H_Q']) / ((data['d2H_Q'] - data['delta_star_d2H']) * data['m_d2H'])

data[['d18O_I', 'd18O_Q', 'E_I_d18O', 'd2H_I', 'd2H_Q', 'E_I_d2H']]
```

2. Estiamtion of Uncertainiteis using Bayesian

```python
import pymc3 as pm

import theano.tensor as tt
```

```python
# Define the data
x1_lakewater = np.array([-2.0, -25.0])  # δ18O and δ2H values for lakewater
s1_inflow = np.array([-5.0, -34.0])  # δ18O and δ2H values for inflowing water
A = np.array([1.7881, 1.5824])  # A values for δ18O and δ2H
D = np.array([5.6063, 0.2946])  # D values for δ18O and δ2H

# Uncertainties
x1_uncertainty = np.array([0.2, 3.0])  # Uncertainty in lakewater measurements
s1_uncertainty = np.array([0.2, 3.0])  # Uncertainty in inflow measurements
A_uncertainty = np.array([0.005, 0.03])  # Uncertainty in A
D_uncertainty = np.array([0.005, 0.03])  # Uncertainty in D

# The Model
with pm.Model() as model:
    # Uniform distribution for E/I ratio
    r = pm.Uniform('r', lower=0, upper=1)

    # Normal distribution for mu0 (inflow water composition), A, and D
    mu0 = pm.Normal('mu0', mu=s1_inflow, sd=s1_uncertainty, shape=2)
    A = pm.Normal('A', mu=A, sd=A_uncertainty, shape=2)
    D = pm.Normal('D', mu=D, sd=D_uncertainty, shape=2)

    # Model equation for final isotope composition of water
    mu = r * A * D + mu0 / (r * A + 1)

    # Likelihood (sampling distribution) of observations
    Y_obs = pm.Normal('Y_obs', mu=mu, sd=x1_uncertainty, observed=x1_lakewater)

    # Using Metropolis Hastings Sampling
    step = pm.Metropolis()

    # Sample from the posterior
    trace = pm.sample(5000, step=step, return_inferencedata=False)

# Plot the trace
pm.traceplot(trace)
```

```python
summary = pm.summary(trace, hdi_prob=0.94) # 94% Highest Posterior Density Interval (HPDI)
print(summary)
```
